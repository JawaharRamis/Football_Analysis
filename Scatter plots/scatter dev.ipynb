```python
fbref = sd.FBref(leagues="ENG-Premier League", seasons=2022)
# df = fbref.read_team_season_stats(stat_type='passing')
df = fbref.read_player_season_stats(stat_type="possession")
df
```

```python
df = df.reset_index()
df = df.drop(columns=['team', 'season', 'league'])
```

```python
players = [
    'Mohamed Salah',
    'Gabriel Jesus',
    'Bukayo Saka',
    'Erling Haaland',
    'Aleksandar Mitrović',
    'Wilfried Zaha',
    'Danny Welbeck',
    'Jarrod Bowen',
    'Harry Kane',
    'Dejan Kulusevski',
    'Ivan Toney',
    'Marcus Rashford',
    'Kevin De Bruyne', 
    'Erling Haaland',
    'Allan Saint-Maximin',
    'Anthony Gordon',
    'Raheem Sterling',
    'Martinelli'
]
```

```python
df['Pos'] = df['Pos'].fillna('XX')
df = df[df['Pos'].str.contains("FW")]
```

```python
df_main = df[~df["player"].isin(players)].reset_index(drop = True)
df_highlight = df[df["player"].isin(players)].reset_index(drop = True)
```

```python
# -- Plot the chart


x_param = df['Touches']['Att Pen']
y_param = df['Carries']['1/3']


fig = plt.figure(figsize = (7,5), dpi = 300)
ax = plt.subplot(facecolor = "#e9edcc")

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.scatter(
    df_main['Touches']['Att Pen'], 
    df_main['Carries']['1/3'], 
    s = 40, 
    alpha = 0.6, 
    color = "#780b68",
    zorder = 3
)

ax.scatter(
    df_highlight['Touches']['Att Pen'], 
    df_highlight['Carries']['1/3'], 
    s = 50, 
    alpha = 0.95, 
    color = "#F64740",
    zorder = 3,
    ec = "#000000",
)


ax.plot(
    [ x_param.median(), x_param.median()],
    [ax.get_ylim()[0], ax.get_ylim()[1]], 
    ls = ":",
    color = "#b82ea5",
    zorder = 2
)

ax.plot(
    [ax.get_xlim()[0], ax.get_xlim()[1]],
    [ y_param.median(),  y_param.median()], 
    ls = ":",
    color = "#b82ea5",
    zorder = 2
)

ax.grid(True, ls = ":", color = "#4e590b")


texts = []

for index, name in enumerate(df_highlight["player"]):
    X = df_highlight['Touches']['Att Pen'].iloc[index]
    Y = df_highlight['Carries']['1/3'].iloc[index]
    texts.append(plt.text(X,Y, name, color ='#c8f7c8', fontweight='medium', size='small'))
#     text_ = ax.annotate(
#         xy = (X, Y),
#         text = name.split(" ")[1],
#         ha = "center",
#         va = "center",
#         xytext = (0, 10),
#         textcoords = "offset points",
#         weight = "bold"
#     )

#     text_.set_path_effects(
#                 [path_effects.Stroke(linewidth=2.5, foreground="white"), 
#                 path_effects.Normal()]
#             )
adjust_text(texts, precision=0.001,
        expand_text=(1.01, 1.05), expand_points=(1.01, 1.05),
        force_text=(0.01, 0.25), force_points=(0.01, 0.25))
ax.set_xlabel("Touches in Penalty box")
ax.set_ylabel("Progressive carries in final third")

ax.spines['bottom'].set_color('#c8f7c8')
ax.spines['left'].set_color('#c8f7c8')
ax.xaxis.label.set_color('#c8f7c8')
ax.yaxis.label.set_color('#c8f7c8')
ax.tick_params(axis='x', colors='#c8f7c8')
ax.tick_params(colors='#c8f7c8')

fig_text(
    x = 0.9, y = 0.9, 
    s = "<Premier League> forwards | Season <%s>\n <Touches in Penalty box> vs <Progressive carries in final third> " %(SEASON),
    highlight_textprops=[{"color":"#F64740", "style":"italic"},
                         {"color":"#F64740", "style":"italic"},
                        {"color":"#F64740", "style":"italic", "fontsize" : "8"}, 
                         {"color":"#F64740", "style":"italic", "fontsize" : "8"}],
    va = "bottom", ha = "right",
    fontsize = 12, color = "#c8f7c8", font = "Karla", weight = "bold"
)

image_ax = fig.add_axes(
        [0.14,0.85, .22, .15],
        fc='#031738', 
    )

logo_dir = '../assets/logos/PL/Premier League.png'
club_icon = Image.open(logo_dir)
# club_icon = np.array(club_icon).astype(np.float) / 255

image_ax.imshow(club_icon)
image_ax.axis('off')


plt.savefig(
	"Viz/scatter_players.png",
	dpi = 600,
	facecolor = "#031738",
	bbox_inches="tight",
    edgecolor="none",
	transparent = True
)
```
