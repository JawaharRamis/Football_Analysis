## BASIC PIZZA CHART - SINGLE PLAYER

```python
import pandas as pd 
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from PIL import Image
import math
from mplsoccer import PyPizza, add_image, FontManager, Sbopen

PLAYER_1 = "Bukayo Saka"
POS ="FW"
player_type ='Wide Player'
```

```python
##Scraped using dbrefscraping.ipynb

df= pd.read_csv(r"C:\Users\jawah\Documents\Football Analysis\Data\fbref_outfield_player_stats_Big-5-European-Leagues_2022-2023_latest.csv")
df = df[df['Pos'].str.contains(POS)] #filter by position
team = df[df['Player'] == PLAYER_1].iloc[0]['Squad']
min_played = df[df['Player'] == PLAYER_1].iloc[0]['Min']
age = df[df['Player'] == PLAYER_1].iloc[0]['Age']


def get_params(player_type):
    # STRIKER TEMPLATE
    if player_type =='Striker':
        TEMPLATE = 'Striker'
        PARAMS = ['Player','SoT%','Sh/90', 'G/Sh', 'npxG', 'KP','PPA', 
                  'Prog', 'SCA',  'Dist ', 'Prog.1']
        DISPLAY_PARAMS = ['Shots on Target %','Shots/90', 'Goals/shot', 'npxG ', 'Key Passes','18 yard box passes',
                         'Progressive passes', 'Shot Creation Actions', 'Shot Average Distance', 'Progressive Passes Received']

    # WINGER TEMPLATE
    elif player_type =='Wide Player':
        TEMPLATE = 'Wide Player'
        PARAMS = ['Player','Sh/90','npxG', 'KP', 'Prog', 'SCA','xA', 
                  'CrsPA', 'Succ%', 'PrgDist', 'Prog.1']
        DISPLAY_PARAMS = ['Shots/90','npxG', 'Key Passes', 'Progressive passes', 'Shot Creation Actions','Expected Assists',
                         'succ. 18 yard crosses', 'succ. dribble %', 'Progressive carry distance', 'Progressive Passes Received']

    # attacking/central midfielder TEMPLATE
    elif player_type =='Central/Attacking Midfielder':
        TEMPLATE = 'Central/Attacking Midfielder'
        PARAMS = ['Player','Sh/90','npxG', 'KP', 'Prog', 'SCA','xA', 
                  'CrsPA', 'Succ%', 'PrgDist', 'Prog.1']
        DISPLAY_PARAMS = ['Shots/90','npxG', 'Key Passes', 'Progressive passes', 'Shot Creation Actions','Expected Assists',
                         'succ. 18 yard crosses', 'succ. dribble %', 'Progressive carry distance', 'Progressive Passes Received']

    # defensive midfielder TEMPLAT
    elif player_type =='Defensive Midfielder':
        TEMPLATE = 'Defensive Midfielder'
        PARAMS = ['Player','Cmp%','Cmp%.3', '1/3', 'Prog', 'Att 3rd.1',
                  'Tkl+Int', '%', 'TklW', 
                  'PrgDist', 'Blocks']
        DISPLAY_PARAMS = ['Pass Completion %','Long Pass Completion %', 'Final 1/3 entry passes', 'Progressive passes', 'Att 3rd Pressures',
                          'Tackles & Interceptions',    'Successful Pressure %', 'Possession winning tackle',
                          'Progressive carry distance', 'Blocks']

    # defender TEMPLATE
    elif player_type =='Defender':
        TEMPLATE = 'Defender'
        PARAMS = ['Player','Att 3rd','Tkl%', 'Att 3rd.1', '%', 'Tkl+Int',
                  'Cmp%', 'Cmp%.3', 'Succ%', 'PrgDist', 'Carries']
        DISPLAY_PARAMS = ['Att 3rd Tackles','Dribbler Tackler %', 'Att 3rd Pressures', 'Successful Pressure %', 'Tackles & Interceptions',
                          'Pass Completion %', 'Long Pass Completion %', 'succ. dribble %', 'Progressive carry distance', 'Carries']
#     goalkeeper TEMPLATE
    elif player_type =='Goalkeeper':
        TEMPLATE = 'Goalkeeper'
        PARAMS = ['Player','Save%','Launch%', 'AvgLen.1', 'Stp%', '#OPA/90']
        DISPLAY_PARAMS = ['Att 3rd Tackles','Passes Launched', 'Avg Length from GK', 'High Claim %', 'Def action outside penalty box']
    return TEMPLATE, PARAMS, DISPLAY_PARAMS

TEMPLATE, PARAMS, DISPLAY_PARAMS = get_params(player_type)

df = df[PARAMS]
df = df.fillna(0)
for column in df.columns[1:]:
        df[column] = df[column].astype(np.float32)
```

```python
plot_params = PARAMS[1:]
player = df.loc[df['Player'] == PLAYER_1].reset_index()
player = list(player.iloc[0])[2:]
```

```python
values = []

for x in range(len(plot_params)):
    values.append(math.floor(stats.percentileofscore(df[plot_params[x]], player[x])))
    
num_params = len(values)
```

```python
baker = PyPizza(
    params= DISPLAY_PARAMS,                  # list of parameters
    straight_line_color="#ffb3f9", # color for straight lines
    background_color="#f4f5e1",
    straight_line_lw=3, # linewidth for straight lines
    last_circle_lw=5,              # linewidth of last circle
    other_circle_lw=3,              # linewidth for other circles
    other_circle_ls="-.",   
    last_circle_color= '#315c36',
    other_circle_color = '#89f596',
)



fig, ax = baker.make_pizza(
    values,              # list of values
    figsize=(12, 12),      # adjust figsize according to your need
    param_location=114, 
    value_bck_colors= ['#ffffff']*num_params,
    blank_alpha = 0.1,
    kwargs_slices=dict(
        facecolor="#82274e", edgecolor="#0e5917",
        zorder=2, linewidth=3
    ),                   # values to be used when plotting slices
    kwargs_params=dict(
        color="#315c36", fontsize=12, va="center"
    ),                   # values to be used when adding parameter
    kwargs_values=dict(
        color="#001100", fontsize=12, zorder=3,
        bbox=dict(
            edgecolor="black", facecolor="#e7f06e",
            boxstyle="round,pad=0.2", lw=1
        )
    )                    # values to be used when adding parameter-values
)


fig.text(
    0.515, 1, "%s(%s)- %s" %(PLAYER_1, age, team), size=25,
    ha="center", color="#750251"
)

# add subtitle
fig.text(
    0.515, 0.96,
    "Percentile Rank vs Top 5 League Forwards | %s Template | Season 2022-23"%(player_type),
    size=12,
    ha="center",  color="#942188"
)

CREDIT_1 = "data: statsbomb viz fbref"
CREDIT_2 = ""

fig.text(
    0.99, 0.02, f"{CREDIT_1}\n{CREDIT_2}", size=9,
    color="#fffefc",
    ha="right"
)

fig.savefig('./viz/%s.png'%(PLAYER_1), dpi=600, bbox_inches='tight', pad_inches=0.5)
```
