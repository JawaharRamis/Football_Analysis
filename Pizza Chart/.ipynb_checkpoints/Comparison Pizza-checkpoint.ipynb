```python
import pandas as pd 
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from PIL import Image
import math
from mplsoccer import PyPizza, add_image, FontManager, Sbopen
from highlight_text import fig_text
```

```python
PLAYER_1 = "Brian Brobbey"
PLAYER_2 = "Sardar Azmoun"
POS ="FW"
```

```python
df= pd.read_csv(r"C:\Users\jawah\Documents\Football Analysis\Data\fbref_outfield_player_stats_Bundesliga_2021-2022_latest.csv")
df = df[df['Pos'].str.contains(POS)] #filter by position
team_1 = df[df['Player'] == PLAYER_1].iloc[0]['Squad']
team_2 = df[df['Player'] == PLAYER_2].iloc[0]['Squad']

params = ['Player','Gls', 'SoT','npxG', 'Dist', 'npxG/Sh','G/Sh', 'SoT%']
display_params = ['Goals', 'Shots on Target ', 'Expected-npxG', 
                  'Avg shot distance', 'Non-Penalty xG/shot', 'Goals/shot', '% of on-target shots']

df=df[params]

df = df.fillna(0)
for column in df.columns[1:]:
        df[column] = df[column].astype(np.float32)
```

```python
plot_params = params[1:]

player = df.loc[df['Player'] == PLAYER_1].reset_index()
player = list(player.iloc[0])[2:]

player_2 = df.loc[df['Player'] == PLAYER_2].reset_index()
player_2 = list(player_2.iloc[0])[2:]
```

```python
values = []
values_2 = []

for x in range(len(plot_params)):
    values.append(math.floor(stats.percentileofscore(df[plot_params[x]], player[x])))
    
for x in range(len(plot_params)):
    values_2.append(math.floor(stats.percentileofscore(df[plot_params[x]], player_2[x])))
    
```

```python
# font_normal = FontManager(("https://github.com/google/fonts/blob/main/apache/roboto/"
#                            "Roboto%5Bwdth,wght%5D.ttf?raw=true"))
# font_italic = FontManager(("https://github.com/google/fonts/blob/main/apache/roboto/"
#                            "Roboto-Italic%5Bwdth,wght%5D.ttf?raw=true"))
# font_bold = FontManager(("https://github.com/google/fonts/blob/main/apache/robotoslab/"
#                          "RobotoSlab%5Bwght%5D.ttf?raw=true"))
```

```python
baker = PyPizza(
    params=display_params,                  # list of parameters
    straight_line_color="#f0f7f7", # color for straight lines
    background_color="#2b2b2a",
    straight_line_lw=3, # linewidth for straight lines
    last_circle_lw=2,               # linewidth of last circle
    other_circle_lw=1,              # linewidth for other circles
    other_circle_ls="-." # linestyle for other circles
)

params_offset = [
    False, True, False, True, True,
    False, False, True, True, True, True
]

fig, ax = baker.make_pizza(
    values,                     # list of values
    compare_values=values_2,    # comparison values
    figsize=(8, 8),             # adjust figsize according to your need
    blank_alpha=0.6,
    param_location = 112, 
    kwargs_slices=dict(
        facecolor="#3FA796", edgecolor="#222222",
        zorder=1, linewidth=2
    ),                          # values to be used when plotting slices
    kwargs_compare=dict(
        facecolor="#FF9300", edgecolor="#222222",
        zorder=3, linewidth=2
    ),
    kwargs_params=dict(
        color="#ffffff", fontsize=12, zorder=5,
#         fontproperties=font_normal.prop,
        va="center"
    ),                          # values to be used when adding parameter
    kwargs_values=dict(
        color="#000000", fontsize=12,
#         fontproperties=font_normal.prop,
        zorder=3,
        bbox=dict(
            edgecolor="#000000", facecolor="cornflowerblue",
            boxstyle="round,pad=0.2", lw=1
        )
    ),                          # values to be used when adding parameter-values labels
    kwargs_compare_values=dict(
        color="#000000", fontsize=12, zorder= 3,
        bbox=dict(edgecolor="#000000", facecolor="#FF9300", boxstyle="round,pad=0.2", lw=1)
    ),                          # values to be used when adding parameter-values labels
)

fig_text(
    0.515, 1.01, "<%s> vs <%s>" %(PLAYER_1, PLAYER_2), size=24, fig=fig,
    highlight_textprops=[{"color": '#1A78CF'}, {"color": '#EE8900'}],
    ha="center", color="#ffffff"
)

fig_text(
    0.515, 0.96, "<%s> | Percentile Rank against %s" %('2021/22', 'Top 5 league forwards under 24 '), size=12, fig=fig,
    ha="center", color="#ffffff"
)


CREDIT_1 = "data: statsbomb viz fbref"
CREDIT_2 = ""

fig.text(
    0.99, 0.02, f"{CREDIT_1}\n{CREDIT_2}", size=9,
    color="#fffefc",
    ha="right"
)

fig.savefig('./viz/%s vs %s.png'%(PLAYER_1, PLAYER_2), dpi=fig.dpi, bbox_inches='tight', pad_inches=0.5)
```
